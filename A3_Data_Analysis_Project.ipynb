### Attributes/Variables

- **pH value:** The pH of water is an essential metric in determining its acid-base balance. It can also indicate if the water is acidic or alkaline. The WHO recommends a pH range of `6.5 to 8.5` as the highest allowable level. The present investigation ranges were `6.52-6.83`, which are within the WHO guideline range.  

- **Hardness:** Hardness was initially described as water's ability to precipitate soap due to `calcium` and `magnesium`. `Calcium` and `magnesium` salts are the primary causes of hardness. The amount of hardness in raw water is determined by the length of time water comes in contact with hardness generating material.

- **Solids (Total dissolved solids - TDS):** A high TDS measurement implies that the water is heavily mineralized. The desirable limit for TDS is `500 mg/l`, while the maximum level for drinking is `1000 mg/l`. 

- **Chloramines:** The two main disinfectants utilized in public water systems are `chlorine` and `chloramine`. Drinking water can include up to `4 mg/L` of chlorine (or 4 ppm), which is regarded as a safe quantity. 

- **Sulfate:** Sulfates are naturally occurring compounds found in minerals, soil, and rocks. The quantity of sulfate in saltwater is around `2,700 milligrams per liter (mg/L)`. In most freshwater supplies, it varies from 3 to `30 mg/L`, while significantly greater amounts `(1000 mg/L)` are found in specific geographical areas.  

- **Conductivity:** Pure water is not a good conductor of electricity. The electrical conductivity of water is improved by an increase in ion concentration. It is typically determined by the amount of dissolved particles present in the water.  

- **Organic Carbon:** `Total Organic Carbon (TOC)` in source waters is derived from both natural organic matter (NOM) breakdown and manmade sources. The total quantity of carbon in organic compounds in pure water is measured as TOC. The US EPA recommends `2 mg/L` TOC in treated / drinking water and `4 mg/Lit` in source water used for treatment. 

- **Trihalomethanes:** THMs are compounds that can be found in chlorine-treated water. THM concentrations in drinking water vary depending on the amount of organic material in the water, the amount of chlorine needed to treat the water, and the temperature of the water being treated. THM levels in drinking water up to `80 ppm` are deemed safe.  

- **Turbidity:** The amount of solid stuff present in the suspended state determines how turbid the water is. The test is used to determine the quality of waste discharge with regard to colloidal matter and measures the light-emitting capabilities of water. The Wondo Genet Campus's mean turbidity value `(0.98 NTU)` is less than the WHO-recommended threshold of `5.00 NTU`. 

- **Potability:** Indicates whether water is safe for human consumption, with `1` indicating potable and `0` indicating not potable.

## Handling Missing Values
  
  * They were __three columns with Null values__.
  * The columns were `ph`, `Sulfate` and `Trihalomethanes`.

````python
def visualise_missing_values():
    plt.figure(figsize=(6,4))
    values_missed = round(WATER_POTABILITY_DATA.isna().sum() * 100 / len(WATER_POTABILITY_DATA), 2)
    values_missed = values_missed[values_missed > 0]
    values_missed.sort_values(inplace=True, ascending=False)
    sns.barplot(x = values_missed.index, y = values_missed.values)
    plt.title('Missing Values')
    plt.show()
````

````python
visualise_missing_values()
````

#### We can fill the missing values with their Mean because features are normally distributed.

````python
# This dictionary stores the lower and upper limits for individual columns before filling and cleaning data
PRE_FILL_LIMITS = {}
# Structure of the dictionary: {"column": (lower_limit, upper_limit)}
# In this function we remove outliers by comparing it with upper and lower limits 
def get_limits(column):
    upper_limit = WATER_POTABILITY_DATA[column].mean() + 3 * WATER_POTABILITY_DATA[column].std()
    lower_limit = WATER_POTABILITY_DATA[column].mean() - 3 * WATER_POTABILITY_DATA[column].std()
    
    PRE_FILL_LIMITS[column] = (lower_limit, upper_limit)
````

````python
for column in WATER_POTABILITY_DATA.columns[0:9]:
    get_limits(column)
    
print(PRE_FILL_LIMITS)
````

### Filling missing values

````python
# Imputing missing values with mean values
def fill_missing_values(column):
    lower_limit = PRE_FILL_LIMITS[column][0]
    upper_limit = PRE_FILL_LIMITS[column][1]
    local_mean = WATER_POTABILITY_DATA[(WATER_POTABILITY_DATA[column] >= lower_limit) & (WATER_POTABILITY_DATA[column] <= upper_limit)][column].mean()
    WATER_POTABILITY_DATA[column].fillna(local_mean, inplace=True)
````

````python
fill_missing_values(PH_COLUMN)
````

### Handling Outliers

An `outlier` is an observation in a dataset that is significantly different from the majority of the other observations. Outliers can arise for a variety of reasons, such as `measurement errors`, `errors in data entry`, or the `presence of rare events` or `patterns in the data`. `Outliers` can have a significant impact on the performance of machine learning algorithms, since they can cause the model to overfit to the noise and random fluctuations in the data, rather than the underlying pattern. Therefore, it is important to identify and handle outliers in a dataset in order to improve the performance of machine learning models.

````python
for column in WATER_POTABILITY_DATA.columns[:9]:
    WATER_POTABILITY_DATA = WATER_POTABILITY_DATA[(WATER_POTABILITY_DATA[column] >= PRE_FILL_LIMITS[column][0]) & (WATER_POTABILITY_DATA[column] <= PRE_FILL_LIMITS[column][1])]
````

Now let's understand the machine learning models used in our analysis: 

### <u>KNN</u>

KNN is an abbreviation for k-nearest neighbours.
It is a supervised learning technique that is used for classification and regression. A KNN algorithm classifies data points based on the majority class of their nearest neighbours. This is accomplished by computing the distance between a particular data point and all other points in the dataset, picking the k-nearest points, and then classifying the majority of those points. KNN is a simple and effective method that may be used for a variety of tasks such as image and text categorization and anomaly detection. It is especially effective when the underlying data is not linearly separable. 

### <u>Decision Tree</u>

Decision trees are supervised learning algorithms that are used for classification and regression. A decision tree is a collection of rules that predicts the value of a target variable based on the values of one or more input variables. These rules are expressed in the form of a tree structure, with the root node representing the complete dataset and the leaves indicating the projected class or value. Decision trees are valuable because they can handle both continuous and categorical data and are simple to analyse. They are also generally unaffected by outliers and missing data. They are, however, prone to overfitting and may underperform on datasets with complicated associations. 

### <u>Random Forest</u>

Random forests are a classification and regression ensemble learning technique. They are known as "random forests" because they are made up of a huge number of separate decision trees, each of which is trained on a different subset of the data.
The projections of each individual tree are then merged to provide a final prediction. This method provides a number of advantages over employing a single decision tree. For example, it eliminates overfitting that can occur with individual trees and can increase the model's forecast accuracy. Random forests are one of the most accurate and resilient machine learning algorithms and are extensively utilised for a number of applications, including image and text categorization.

## Data Scaling - Standardization

The scaling of features is an important stage in modelling algorithms using datasets. The data that is often utilised for modelling is obtained through a variety of ways, including: Questionnaire, polls, surveys. Hence, the acquired data comprises characteristics of various dimensions and sizes. Different sizes of data characteristics have a negative impact on dataset modelling. As a result, the accuracy and misclassification error rates of the prediction are skewed. Therefore, the data must be scaling before modelling. 

Standardization is a scaling technique that converts the statistical distribution of the data. It scales the distribution to have a mean of zero and a standard deviation of one by dividing by the standard deviation, subtracting the mean (a process known as centering), and scaling each input variable independently. 

The standard score of a sample x is calculated as: <mark> z = (x - u) / s </mark> where s equals "standard deviation" and u equals "mean"
